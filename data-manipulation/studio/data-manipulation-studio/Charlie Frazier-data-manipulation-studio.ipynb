## Clean Your Data

In the last lesson, we cleaned the data related to San Francisco. Pull up your notebook from the last lesson and use it as a reference to clean up these new dataframes.

```python
# Clean your data here!

import pandas as pd
import os

print("Working Directory:", os.getcwd())

try:
    
    baltimore_data = pd.read_csv("dataset/baltimore_9-24-2016_9-30-2017.csv")
    print(baltimore_data.head())
    print(baltimore_data.info())
    
    
    baltimore_data.dropna(inplace=True)
    baltimore_data.drop_duplicates(inplace=True)
    
    if 'date' in baltimore_data.columns:
        baltimore_data['date'] = pd.to_datetime(baltimore_data['date'])
    else:
        print("Baltimore data.")

    if 'old_name' in baltimore_data.columns:
        baltimore_data.rename(columns={'old_name': 'new_name'}, inplace=True)
    else:
        print("renaming column.")

    print(baltimore_data.describe())

    def clean_data(dataframe):
        dataframe.dropna(inplace=True)
        dataframe.drop_duplicates(inplace=True)
        return dataframe

    boston_data = clean_data(pd.read_csv("dataset/boston_9-24-2016_9-30-2017.csv"))
    new_york_data = clean_data(pd.read_csv("dataset/new-york_9-24-2016_9-30-2017.csv"))
    philadelphia_data = clean_data(pd.read_csv("dataset/philadelphia_9-24-2016_9-30-2017.csv"))

    baltimore_data.to_csv("cleaned_baltimore_data.csv", index=False)

except FileNotFoundError as e:
    print(f"File not found: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of **unit of sale** in the Northeast region? 
2. For each region, what is the average number of pumpkins per variety that came into terminal markets for the year? 

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.
data = {
    'region': ['Northeast', 'Northeast', 'Northeast', 'Northeast'],
    'variety': ['Small', 'Large', 'Small', 'Large'],
    'unit_of_sale': ['Per Pound', 'Per Pound', 'Per Pumpkin', 'Per Pumpkin'],
    'low_price': [0.50, 0.75, 1.00, 1.25],
    'high_price': [0.60, 0.85, 1.10, 1.35],
    'num_pumpkins': [100, 150, 200, 180]}

northeast_data = pd.DataFrame(data)

# Question #1
mean_prices = northeast_data.groupby('unit_of_sale').agg(
    mean_low_price=('low_price', 'mean'),
    mean_high_price=('high_price', 'mean'))

print(mean_prices)
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.

data = {
    'region': ['Northeast', 'Northeast', 'Northeast', 'Northeast'],
    'variety': ['Small', 'Large', 'Small', 'Large'],
    'unit_of_sale': ['Per Pound', 'Per Pound', 'Per Pumpkin', 'Per Pumpkin'],
    'low_price': [0.50, 0.75, 1.00, 1.25],
    'high_price': [0.60, 0.85, 1.10, 1.35],
    'num_pumpkins': [100, 150, 200, 180]}

northeast_data = pd.DataFrame(data)

# Question #2
avg_pumpkins_per_variety = northeast_data.groupby(['region', 'variety']).agg(
    average_num_pumpkins=('num_pumpkins', 'mean'))

print(mean_prices, avg_pumpkins_per_variety)
```
