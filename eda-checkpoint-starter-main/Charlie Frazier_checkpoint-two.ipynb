```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
hr_data_plot = pd.read_csv(r"C:\Users\charl\Desktop\LaunchCode-Data-Analysis\Data-Analysis-Projects\eda-checkpoint-starter-main\Human_Resources_Data.csv")

# Create a numeric binary indicator that equates 'yes' to 1 and 'no' to 0
hr_data_plot['left'] = hr_data_plot['left'].map({'yes': 1, 'no': 0})

# Group by 'satisfaction_level' and aggregate by the mean of 'left' and the average 'time_spend_company'
grouped_data = hr_data_plot.groupby('satisfaction_level').agg({
    'left': 'mean',  
    'time_spend_company': 'mean' 
}).reset_index()

# Dual-Axis Plot chart
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.set_xlabel('Satisfaction Level')
ax1.set_ylabel('Average Number of People Who Left', color='tab:red')
ax1.plot(grouped_data['satisfaction_level'], grouped_data['left'], color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx() 
ax2.set_ylabel('Time Spent at Company (Years)', color='tab:blue')  
ax2.plot(grouped_data['satisfaction_level'], grouped_data['time_spend_company'], color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Correlation between Satisfaction Level, Departure Rate, and Time Spent at Company')
plt.grid(True)
plt.show()
```
