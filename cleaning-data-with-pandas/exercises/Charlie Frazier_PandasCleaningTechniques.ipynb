## Irregular Data

With missing data out of the way, turn your attention to any outliers. Just as we did for missing data, we first need to detect the outliers.

```python
# Keep an eye out for outliers!

import pandas as pd

womens_clothing_ecommerce = pd.read_csv(r"/Users/charl/Desktop/LaunchCode-Data-Analysis/Data-Analysis-Projects/cleaning-data-with-pandas/exercises/Womens-Clothing-Ecommerce.csv")

irregular_data = womens_clothing_ecommerce['Positive Feedback Count']

Q1 = irregular_data.quantile(0.25)
Q3 = irregular_data.quantile(0.75)
IQR = Q3 - Q1
lower_quantile = Q1 - 1.5 * IQR
upper_quantile = Q3 + 1.5 * IQR

outliers = womens_clothing_ecommerce[(irregular_data < lower_quantile) | (irregular_data > upper_quantile)]

print("Outliers Identified:")
print(outliers[['Positive Feedback Count']])








```

## Unnecessary Data

## Inconsistent Data

Inconsistent data is likely due to inconsistent formatting and can be addressed by re-formatting all values in a column or row.

```python
# Look out for inconsistent data!

import pandas as pd

womens_clothing_ecommerce = pd.read_csv(r"/Users/charl/Desktop/LaunchCode-Data-Analysis/Data-Analysis-Projects/cleaning-data-with-pandas/exercises/Womens-Clothing-Ecommerce.csv")

womens_clothing_ecommerce['Category'] = womens_clothing_ecommerce['Category'].str.lower().str.strip()

corrections = {'dreses': 'dresses', 'intimate': 'intimates', 'accesories': 'accessories'}
womens_clothing_ecommerce['Category'] = womens_clothing_ecommerce['Category'].replace(corrections)

womens_clothing_ecommerce['Price'] = pd.to_datetime(['Date'], errors = 'coerce')
womens_clothing_ecommerce['Price'] = pd.to_numeric(womens_clothing_ecommerce['Price'].str.replace('$', ''), errors='coerce')

womens_clothing_ecommerce.to_csv('Cleaned Data', index=False)

print("Dataset Cleaned")

```
